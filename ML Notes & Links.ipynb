#  Web Scrapping Code


```python
#https://stackabuse.com/text-summarization-with-nltk-in-python/
```

```python
import bs4 as bs
import urllib.request
import re

scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/India')
article = scraped_data.read()

parsed_article = bs.BeautifulSoup(article,'lxml')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    #print(p.text)
    article_text += p.text
article_text 
```
